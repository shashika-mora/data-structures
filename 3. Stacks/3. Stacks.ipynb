# 3 ðŸ¥ž Stacks (LIFO)

A **Stack** is a linear data structure that follows the **LIFO (Last-In, First-Out)** principle. Where elements are inserted and removed from a single end known as the `top`. It plays a crucial role in areas such as function call management, expression evaluation, and backtracking algorithms.

**Key Topics Covered:**
* **The Mental Model:** LIFO (Stack of Plates).
* **Implementation:** Using Python Lists correctly.
* **System Architecture:** The CPU Call Stack.
* **Algorithms:** Undo, Syntax Parsing, and DFS.

> Arrays are **static** data structures, while Stacks are **dynamic** data structures in sense that they can grow and shrink in size.

## 3.1 ðŸ¥ž The Mental Model (LIFO)

Imagine a stack of plates in a cafeteria. 

1.  **Push:** You place a new plate on **top**.
2.  **Pop:** You can only take the plate from the **top**.
3.  **Peek:** You can see the top plate without removing it.

Additionally we use:

* isEmpty: Check if the stack is empty.
* isFull: Check if the stack is full.
* peek: Get the top element.
* size: Get the number of elements in the stack.

You cannot access the bottom plate without removing all the ones above it.

```python
from typing import Any, Optional, List

class Stack:
    """LIFO Stack implementation using a Python List."""
    def __init__(self):
        # We use a private list. We restrict access so you CANNOT say stack[0]. You must use pop().
        self._items: List[Any] = []

    def is_empty(self) -> bool:
        """Checks if Stack is Empty. O(1)."""
        return len(self._items) == 0

    def push(self, item: Any):
        """Adds to Top. O(1)."""
        self._items.append(item)

    def pop(self) -> Optional[Any]:
        """Removes from Top. O(1)."""
        if self.is_empty():
            return None
        return self._items.pop()

    def peek(self) -> Optional[Any]:
        """Looks at Top. O(1)."""
        if self.is_empty():
            return None
        return self._items[-1]
        # using negetive indexing

    def size(self) -> int:
        """Returns the number of items in the stack. O(1)."""
        return len(self._items)

# --- Test ---
s = Stack()
s.push("Page 1 (Home)")
s.push("Page 2 (Settings)")
s.push("Page 3 (Profile)")

print(f"Currently viewing: {s.peek()}")
print(f"Back button pressed: {s.pop()}")
print(f"Now viewing: {s.peek()}")
```

* A **Linked List** can also be used to implement a stack where each node points to the next node in the stack. (Java)
* The **Top** of the stack is the last node in the linked list.
* The **Bottom** of the stack is the first node in the linked list.

* A **Stack** can also be implemented using a **Linked List** in Python where elements are stored in a sequential manner.
* An **Index** (counter) is used to keep track of the top of the stack.
* **Underflow**(popping from an empty stack) is managed using exception handling.

## 3.2 âš™ï¸ The Engineering Reality: The Call Stack

Why are Stacks important? Because **your computer runs on one**.

When you call a function, the CPU "pushes" the function's local variables and return address onto the **Call Stack**. When the function returns, it "pops" that frame off memory. 

This is why infinite recursion causes a **Stack Overflow** error: you filled up the stack memory limit.

```python
import sys

def recursive_function(counter):
    print(f"Pushing frame {counter} onto the stack...")
    recursive_function(counter + 1)

try:
    recursive_function(1)
except RecursionError:
    print("\nðŸ’¥ STACK OVERFLOW! The stack is full.")
```

## 3.3 ðŸŒ Real-World System Map

Where are Stacks used?

### 1. The Browser "Back" Button
*   **Example:** **Chrome/Firefox History**.
*   **Why?** As you browse `Home -> Search -> Video`, these URLs are pushed onto a stack. When you click "Back", the browser `pops` Video off to show you Search. This is famously a LIFO behavior.

### 2. Syntax Parsing (Compilers)
*   **Example:** **VS Code Bracket Matching**.
*   **Why?** How does Python know if you missed a closing parenthesis `)`? It uses a stack. When it sees `(`, it pushes it. When it sees `)`, it attempts to `pop`. If the stack is empty (nothing to pop) or the popped item doesn't match, it's a Syntax Error.



### 3. Undo/Redo Features
*   **Example:** **Text Editors / Photoshop**.
*   **Why?** Your actions are pushed onto an "Action Stack". `Ctrl+Z` pops the last action and reverses it. `Ctrl+Y` pushes it back.

### 4. Evaluating Postfix Expressions

* Postfix notation (Reverse Polish Notation or RPN) is a mathematical expression where the operator follows its operands. For example: 

    - (8-3)*(5+6)   |         Read the expression from left to right
    - 8 3 - 5 6 + * |      Push operands to the stack

* On encountering an operator, pop the required number of operands from the stack, perform the operation, and push the result back to the stack.

    - 8 3 - 5 6 + * |       Pop 8 and 3, perform 8-3=5, push 5 to stack
    - 5 5 +         |     Pop 5 and 5, perform 5+5=10, push 10 to stack
    - 10 *          |      Pop 10 and 6, perform 10*6=60, push 60 to stack
    - 60            |      Result is 60

* Example: 9 2 A 5 8 4 D S M

    - 9 2 A 5 8 4 D S M | Push 9, 2 to stack and perform 9+2=11, push 11 to stack
    - 11 5 8 4 D S M    | Push 5, 8, 4 to stack and perform 8/4 = 2, push 2 to stack
    - 11 5 2 S M        | Pop 2, 5 and perform 5-2=3, push 3 to stack
    - 11 3 M            | Pop 3, 11 and perform 11*3=33, push 33 to stack
    - 33                | Result is 33
