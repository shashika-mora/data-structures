# 8.6 ðŸŒ³ Trees (Hierarchical Data)

Unlike Arrays and Linked Lists (Linear), **Trees** are **Hierarchical**. They model relationships like File Systems, HTML Logic, and Organizational Charts.

**Key Topics Covered:**
* **Anatomy:** Root, Node, Leaf, Height.
* **The BST Rule:** Left < Root < Right.
* **Algorithms:** recursive Insert and Search ($O(\log n)$).
* **Traversals:** DFS Strategies (In-order, Pre-order, Post-order).

## 1. ðŸ§¬ Anatomy of a Tree

A Tree consists of **Nodes** connected by **Edges**. 

-   **Root:** The top node (No parent).
-   **Leaf:** A node with no children.
-   **Height:** The number of edges from the Root to the furthest Leaf.
-   **Binary Tree:** Every node has at most 2 children (Left/Right).

````python
from typing import Optional

class Node:
    """A single node in a Binary Tree."""
    def __init__(self, key: int):
        self.val = key
        self.left: Optional['Node'] = None
        self.right: Optional['Node'] = None

# Manual Construction (Just to see structure)
root = Node(10)
root.left = Node(5)
root.right = Node(15)

print(f"Root: {root.val}")
print(f"Left Child: {root.left.val}")
````

## 2. ðŸŒ² The Binary Search Tree (BST)

This is the specific type of tree used for fast lookups. 

**The Golden Rule:**
For every node:
1.  Everything in the **Left** subtree is **Smaller**.
2.  Everything in the **Right** subtree is **Larger**.

This allows us to cut the search space in half at every step ($O(\log n)$), just like Binary Search on an array.

````python
class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, key: int):
        """Public wrapper for insert."""
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert_recursive(self.root, key)

    def _insert_recursive(self, current: Node, key: int):
        """Recursive logic to find the correct spot."""
        if key < current.val:
            # Go Left
            if current.left is None:
                current.left = Node(key)
            else:
                self._insert_recursive(current.left, key)
        else:
            # Go Right
            if current.right is None:
                current.right = Node(key)
            else:
                self._insert_recursive(current.right, key)

    def search(self, key: int) -> bool:
        """Returns True if key exists."""
        return self._search_recursive(self.root, key)

    def _search_recursive(self, current: Optional[Node], key: int) -> bool:
        # Base Cases
        if current is None:
            return False # Hit dead end
        if current.val == key:
            return True # Found it!
        
        # Recursive Step
        if key < current.val:
            return self._search_recursive(current.left, key)
        else:
            return self._search_recursive(current.right, key)

# --- Test ---
bst = BinarySearchTree()
bst.insert(10)
bst.insert(5)
bst.insert(15)
bst.insert(2)

print(f"Found 15? {bst.search(15)}")
print(f"Found 99? {bst.search(99)}")
````

## 3. ðŸ•µï¸ Tree Traversals (DFS)

How do we "print" a tree? There are 3 main ways (Depth-First Search).

1.  **In-Order (Left, Root, Right):** Returns sorted data (2, 5, 10, 15).
2.  **Pre-Order (Root, Left, Right):** Good for copying trees.
3.  **Post-Order (Left, Right, Root):** Good for deleting trees (delete children before parent).

````python
def inorder_print(node: Optional[Node]):
    if node:
        inorder_print(node.left)
        print(node.val, end=" ")
        inorder_print(node.right)

print("Sorted Data:")
inorder_print(bst.root)
````

## îžæ½® Mini-Challenge: The File System

Linux treats files as a Tree. 
**Task:** Create a `FileNode` class that has a `name` and a list of `children` (instead of just left/right). 
Build this structure manually:

````python
class FileNode:
    def __init__(self, name):
        self.name = name
        self.children = []

    def add_child(self, node):
        self.children.append(node)

# Build it
root = FileNode("/")
home = FileNode("home")
bin_dir = FileNode("bin")

root.add_child(home)
root.add_child(bin_dir)

home.add_child(FileNode("user"))
home.add_child(FileNode("music"))

# Visualize (Pre-order logic)
def print_tree(node, level=0):
    print("  " * level + f"ðŸ“„ {node.name}")
    for child in node.children:
        print_tree(child, level + 1)

print_tree(root)
````

## ðŸŒŸ Core Insight for Your CSE Career

### 1. The "Unbalanced" Trap
If you insert `1, 2, 3, 4, 5` into a BST, it becomes a straight line to the right. The Tree becomes a **Linked List**. 
-   Balanced Tree Search: $O(\log n)$
-   Unbalanced Tree Search: $O(n)$ (Slow)

In production (DBs), we use **Self-Balancing Trees** (AVL Trees, Red-Black Trees) that automatically rotate nodes to stay balanced.

### 2. The DOM is a Tree
Every website is a Tree (The Document Object Model). `<html>` is the root, `<body>` is a child, `<div>` is a grandchild. When you learn React/Frontend, you are just manipulating a Tree.
