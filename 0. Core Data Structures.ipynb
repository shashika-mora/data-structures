## 0.1 ðŸ§  The Memory Model: Contiguous vs. Scattered

All data structures are just different ways of organizing RAM.

### 1. Contiguous Memory (Arrays)
Data is stored in blocks right next to each other. 
-   **Pros:** CPU Cache friendly (super fast reading).
-   **Cons:** Fixed size. Growing requires copying *everything* to a new, bigger spot.

### 2. Scattered Memory (Linked Lists / Trees)
Data is scattered all over RAM. Each piece holds a "Pointer" (address) to the next piece. 
-   **Pros:** Infinite flexibility. Grow/shrink instantly.
-   **Cons:** "Pointer Chasing" causes CPU Cache misses (slower).

Every structure optimizes for one thing at the expense of another.

| Structure | Strength ðŸ’ª | Weakness ðŸ”» | Access | Search | Insert/Delete |
| :--- | :--- | :--- | :---: | :---: | :---: |
| **Array** | Read Speed | Fixed Size | $O(1)$ | $O(n)$ | $O(n)$ |
| **Linked List** | Flexible Size | Random Access | $O(n)$ | $O(n)$ | $O(1)^*$ |
| **Stack** | LIFO Logic | Search | $O(n)$ | $O(n)$ | $O(1)$ (Top) |
| **Queue** | FIFO Logic | Search | $O(n)$ | $O(n)$ | $O(1)$ (Ends) |
| **Hash Table** | **Speed King** | Unordered | $O(1)$ | $O(1)$ | $O(1)$ |
| **BST Tree** | Sorted Data | Balancing | $O(\log n)$ | $O(\log n)$ | $O(\log n)$ |

 ** $O(1)$ insert for Linked List assumes you already have the pointer to the location.**

## 0.3 ðŸŒ Real-World System Map

Where do you actually use these in a career? Use this map to answer system design interview questions.

### 1. Arrays
*   **Example:** **Screen Pixels (Image Processing)**.
*   **Why?** A 4K screen has 8 million pixels. To change the color of a specific pixel at `(x=1000, y=500)`, the CPU needs to calculate the exact memory address instantly. Since arrays are contiguous blocks, `Address = Start + (y * width + x)`. This is math, not searching. $O(1)$ access.

### 2. Linked Lists
*   **Example:** **Music Playlist (Spotify)**.
*   **Why?** If you want to move song #50 to position #2, an Array would need to shift 48 other songs down in memory. A Linked List just changes two pointers (disconnect #50, rewire it to point to #3). Re-ordering is $O(1)$.

### 3. Stacks (LIFO - Last In, First Out)
*   **Example:** **Browser "Back" Button**.
*   **Why?** As you visit pages A -> B -> C, they are pushed onto a stack. When you click "Back", the browser pops C off the top to show you B. The last page you visited is always the first one you want to return to.
*   **Example:** **Undo/Redo** in text editors.

### 4. Queues (FIFO - First In, First Out)
*   **Example:** **Printer Job Spool**.
*   **Why?** If 5 people send documents to the printer, fairness matters. The first document received must be the first one printed. A Stack (LIFO) would be unfair (the last person would cut the line).
*   **Example:** **Web Server Requests** (handling thousands of users in order).

### 5. Hash Tables
*   **Example:** **Database Indices / User Login**.
*   **Why?** When you log in with username "Alice", the system doesn't search 1 billion users alphabetically. It runs `hash("Alice")` to get a memory address and jumps straight there. Lookup is $O(1)$, independent of database size.

### 6. Trees
*   **Example:** **HTML DOM (Webpages)**.
*   **Why?** Webpages are hierarchical. `<body>` contains `<div>`, which contains `<p>`. If you delete a `<div>`, all its children (`<p>`) are deleted too. Trees model this Parent-Child relationship perfectly.
