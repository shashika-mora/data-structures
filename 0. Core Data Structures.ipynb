## 8.1 ðŸ§  The Memory Model: Contiguous vs. Scattered

All data structures are just different ways of organizing RAM.

### 1. Contiguous Memory (Arrays)
Data is stored in blocks right next to each other. 
-   **Pros:** CPU Cache friendly (super fast reading).
-   **Cons:** Fixed size. Growing requires copying *everything* to a new, bigger spot.

### 2. Scattered Memory (Linked Lists / Trees)
Data is scattered all over RAM. Each piece holds a "Pointer" (address) to the next piece. 
-   **Pros:** Infinite flexibility. Grow/shrink instantly.
-   **Cons:** "Pointer Chasing" causes CPU Cache misses (slower).

## 8.2 ðŸ“Š The Big-O Trade-offs Cheat Sheet

Every structure optimizes for one thing at the expense of another.

| Structure | Strength ðŸ’ª | Weakness ðŸ”» | Access | Search | Insert/Delete |
| :--- | :--- | :--- | :---: | :---: | :---: |
| **Array** | Read Speed | Fixed Size | $O(1)$ | $O(n)$ | $O(n)$ |
| **Linked List** | Flexible Size | Random Access | $O(n)$ | $O(n)$ | $O(1)^*$ |
| **Stack** | LIFO Logic | Search | $O(n)$ | $O(n)$ | $O(1)$ (Top) |
| **Queue** | FIFO Logic | Search | $O(n)$ | $O(n)$ | $O(1)$ (Ends) |
| **Hash Table** | **Speed King** | Unordered | $O(1)$ | $O(1)$ | $O(1)$ |
| **BST Tree** | Sorted Data | Balancing | $O(\log n)$ | $O(\log n)$ | $O(\log n)$ |

 ** $O(1)$ insert for Linked List assumes you already have the pointer to the location.**
