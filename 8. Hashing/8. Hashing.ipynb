## 8.1 ðŸ§® The Hash Function

A **Hash Function** takes an input (Key) and returns a fixed-size integer (Hash Code).

**The Golden Rules:**
1.  **Deterministic:** `hash("cat")` must ALWAYS return the same number.
2.  **Fast:** Must be $O(1)$ to compute.
3.  **Uniform Distribution:** Should spread keys evenly across the table to minimize collisions.

Let's write a simple (bad) one and a better one.

```python
def bad_hash(key_str):
    """Naive hash: Sum of ASCII values."""
    return sum(ord(c) for c in key_str)

def better_hash(key_str):
    """Polynomial Rolling Hash (Used by Python/Java)."""
    hash_val = 0
    p = 31 # Prime number multiplier
    m = 10**9 + 9 # Large prime modulus to keep numbers small
    
    for i, char in enumerate(key_str):
        hash_val = (hash_val + ord(char) * (p ** i)) % m
    return hash_val

print(f"'cat' (Bad): {bad_hash('cat')}")
print(f"'act' (Bad): {bad_hash('act')} <- COLLISION! Same ASCII sum.")
print(f"'cat' (Better): {better_hash('cat')}")
print(f"'act' (Better): {better_hash('act')} <- Unique!")
```

## 8.2 ðŸ’¥ Collisions

Even with a perfect hash function, according to the **Pigeonhole Principle**, if you have more items than table slots, collisions are mathematically inevitable.

### Strategy 1: Chaining (Linked Lists)
The table slot doesn't hold the value; it holds a **Pointer to a Linked List**. If multiple keys hash to index `5`, we just append them to the list at index `5`.
-   **Pros:** Simple, table never "fills up".
-   **Cons:** Cache misses (pointer chasing).

### Strategy 2: Open Addressing (Linear Probing)
If index `5` is taken, try `6`. If `6` is taken, try `7`.
-   **Pros:** Cache friendly (everything is in the array).
-   **Cons:** Sensitivity to clustering (if a block gets full, performance tanks).

## 8.3 ðŸŒ Real-World System Map

Where is Hashing used besides generic Dictionaries?

### 1. Password Storage (Security)
*   **Example:** **User Registration**.
*   **Why?** You NEVER store passwords as plain text. If hackers steal the DB, they see "password123". instead, you store `hash("password123")`. When a user logs in, you hash their input and compare it to the stored hash. You cannot verify the password, only the hash.

### 2. Distributed Caching (Consistent Hashing)
*   **Example:** **Scalable Web Servers**.
*   **Why?** If you have 10 servers, you can determine which server holds user data by `hash(user_id) % 10`. This balances the load evenly across all servers.

### 3. Data Integrity
*   **Example:** **Git**.
*   **Why?** Git identifies files not by name, but by the SHA-1 hash of their contents. If you change ONE character in a file, the hash changes completely. This guarantees that file history is immutable and tamper-proof.

## ðŸŒŸ Core Insight for Your CSE Career

### The Trade-off
Hashing is **Fast** ($O(1)$) but **Unordered**. You cannot print a Hash Table "alphabetically" without sorting it first ($O(N \log N)$). If you need speed AND order, you need a **Tree** (TreeMap in Java), which is $O(\log N)$.
